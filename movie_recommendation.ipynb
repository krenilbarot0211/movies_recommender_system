```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lower case
  text= str(text).lower()
  #panctuation
  text=re.sub(r'[^a-zA-Z]',' ',text)

  words=text.split()
  #lemmatizer
  words=[lemmatizer.lemmatize(word) for word in words if word not in stop_words]
  return " ".join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df=df.reset_index(drop=True)
```

```python
indicies=pd.Series(df.index,index=df['title']).drop_duplicates()
indicies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Example: if your dataframe is called df and movie titles are in df['title']

indices = pd.Series(df.index, index=df['title']).drop_duplicates()
def recommend(title,n=10):
    
    if title not in indices:
        return("movie not found")
    idx=indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
    
```

```python
recommend('Avatar',10)
```
